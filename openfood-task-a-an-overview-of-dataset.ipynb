#### Columns by Suffix per_portion 

 Use ".endswith" within a list of strings in Python
 
"Check endswith with a string. Function **e.split()** gives a list.
Iterate through this list and **check endswith** with each item in the list.

```python
#https://stackoverflow.com/questions/52347490/how-to-use-endswith-within-a-list-of-strings-in-python
#answered  by Austin Sep 15, 2018 at 18:19
#Columns by Suffix per_portion

suffix_list = []

for e in df:
    for element in e.split():
        if element.endswith("per_portion"):
            suffix_list.append(element)

print(suffix_list)
```

#### Columns by Suffix per_hundred 

```python
#https://stackoverflow.com/questions/52347490/how-to-use-endswith-within-a-list-of-strings-in-python

suffix_list = []

for e in df:
    for element in e.split():
        if element.endswith("per_hundred"):
            suffix_list.append(element)

print(suffix_list)
```

#### Columns by Suffix unit

```python
#https://stackoverflow.com/questions/52347490/how-to-use-endswith-within-a-list-of-strings-in-python

suffix_list = []

for e in df:
    for element in e.split():
        if element.endswith("unit"):
            suffix_list.append(element)

print(suffix_list)
```

### Create other_cols list (remaining cols without suffix)

Since I wasn't able to create other_cols with the snippet above, I tried to use filter() function and regex $ to deliver the column with the remaining features that don't have suffix.

In fact, **I didn't need to write many of the snippets above**, cause I'd be able to make it just with the snippet below. 

**Python IF with NOT Operator**

"The main purpose of using if with not is to conditionally execute code based on the absence of a specific value."

"We can use if with logical not operator in Python. The main use of the logical not operator is that it is used to inverse the value. With the help of not operator, we can convert true value to false and vice versa. When not applied to the value so it reverses it and then the final value is evaluated by the if condition. So according to its final value the if block or else block is executed."

https://www.geeksforgeeks.org/python-if-with-not-operator/


```python
#Create lists for each suffix with: end $ using **Pandas filter
#Attention: the first underscore doesn't enter on the code below.

per_hundred = df.filter(regex='per_hundred$', axis=1).columns.tolist()
per_portion = df.filter(regex='per_portion$', axis=1).columns.tolist()
unit = df.filter(regex='unit$', axis=1).columns.tolist()


##Combine all the suffix columns into one list
cols_with_suffix = per_hundred + per_portion + unit

##other_cols is the list of all columns without the specified suffixes
#IF with NOT Operator
#Using IF with NOT, conditionally execute code based on the absence of a specific value.


other_cols = [col for col in df.columns if col not in cols_with_suffix]
other_cols
```
