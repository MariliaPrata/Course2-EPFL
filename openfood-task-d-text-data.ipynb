#### Helper function to preprocess the text

This function preprocesses raw samples of text:

Converts to lowercase
    
Replaces common punctuation marks with whitespace
 
Removes stop words
   
Splits text on whitespace

```python
# Define preprocess_text function

def preprocess_text(sample):
    
    '''
    This function preprocesses raw samples of text:
    - Converts to lowercase
    - Replaces common punctuation marks with whitespace
    - Removes stop words
    - Splits text on whitespace
    
    INPUT: 
    - text: "raw" text (string)
    
    OUTPUT:
    - processed_sample_tokens: list of tokens (list of strings)
    '''
    
    # Convert to lowercase
    clean_sample = sample.lower()
    
    # Replace common punctuation marks with whitespace
    clean_sample = re.sub(r"[,.()]"," ", clean_sample)
    
    # Replace single hyphens with whitespace
    clean_sample = re.sub(r"\s-\s"," ", clean_sample)
    
    # Remove stop words and split on whitespace
    processed_sample_tokens = [tok for tok in clean_sample.split() if tok not in stopwords_english]
    
    return processed_sample_tokens
```

### Collocations

"There is one more thing we can do with our texts. We can look for combinations of words that frequently appear together. We often call these word combinations: phrases or collocations. These word combinations appear in texts more frequently than their constituent words appear individually.

"We will use Gensim’s Phrases model to detect collocations in our corpus. model.phrases will identify the most common collocations and join the constituent tokens into a single token, using the “_” glue character."

```python
# Import Gensim and Phrases

from gensim.models import Phrases
```

####  Find common collocations in ingredients_en

```python
# Find common collocations in ingredients_en

# Create an instance of the Phrases model
n_gram_model = Phrases(preprocessed_ingredients_en,       # Expected input format: list of tokenized texts
                       min_count=5,                       # Keeps only collocations with at least min_counts 
                       delimiter='_') #Original is delimiter=b'_' With b we have error                  # Glue character used to join collocation tokens
```

####  Apply the n_gram_model Model to the ingredients_en lists.

"An n-gram is a sequence of n adjacent symbols in particular order. The symbols may be n adjacent letters (including punctuation marks and blanks), syllables, or rarely whole words found in a language dataset."

```python
# Apply Phrases to ingredients_en

preprocessed_ingredients_en_collocations = [n_gram_model[doc] for doc in preprocessed_ingredients_en]
```

```python
# Add processed press releases (with collocations) to dataframe

df['clean_ingredients_en'] = preprocessed_ingredients_en_collocations
```

### Tf-idf weighting model

"We may want to diminish the importance of widespread words over the entire corpus of documents as they could shadow rarer yet more interesting words. Focus more on rare words across the corpus, as they could be more informative about the content of ingredients_en."

```python
# Import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Create an instance of the tf-idf model
tfidf = TfidfVectorizer()
```

### Use the join() function to join lists of tokens into single strings.

```python
# Join lists of tokens into single strings (for each press release)

X = [" ".join([w for w in doc]) for doc in preprocessed_ingredients_en_collocations]
```

### Fit Method

```python
#we call the fit method of TfidfVectorizer() on our text data.

# Fit model on text data
tfidf.fit(X);
```

### Vocabulary (Unique words in our Corpus)

"The vocabulary is the list of unique words in our corpus. We can check which words are part of the vocabulary using the **get_feature_names_out()** method."

```python
# Show vocabulary
vocabulary = tfidf.get_feature_names_out()

# Print vocabulary length
print('Vocabulary length: {}\n'.format(len(vocabulary)))

# Show a few words from vocabulary
print('Show a few words from vocabulary:')
print(vocabulary[0:500:10])
```

### Transform each sample of text in corpus into its tf-idf representation

"Transform each sample of text in our corpus into its tf-idf representation, we call the transform() method."

```python
# Get tf-idf representation of text data
X_tfidf = tfidf.transform(X)
print(X_tfidf.shape)
```

#### Check what X_tfidf looks like using the following code snippet

```python
df_tfidf = pd.DataFrame(X_tfidf.toarray(),
                        columns=vocabulary)
df_tfidf.head()
```

###  Top words according to tf-idf scores

Plot the top 50 words according to the tf-idf scores.

```python
import random

idx = random.randint(0,len(df)) #Each time a different document with random

# Set number of top words to plot
N = 50

print('Printing top words according to tf-idf for document {}.'.format(idx))
```

```python
# Sort indexes according to tf-idf scores for document idx

index_sorted = np.argsort(df_tfidf.iloc[idx,:])[::-1]  # descending order

# Print sorted indexes according to tf-idf
index_sorted.values
```

```python
# Sort words in vocabulary according to tf-idf

features_sorted = np.array(vocabulary)[index_sorted][0:N]

# Print top words according to tf-idf scores
print(features_sorted)
```

```python
# Sort tf-idf scores (descending order)

#ser.iloc[pos]
#tfidf_weights_sorted = df_tfidf.iloc[idx,:][index_sorted][0:N]

tfidf_weights_sorted = df_tfidf.iloc[idx,:][index_sorted][0:N]

print(tfidf_weights_sorted)
```

### Visualization top words according to tf-idf scores

```python
# Plots top words according to tf-idf scores

# Barplot and font specifications
barplot_specs   = {"color": "mediumpurple", "alpha": 0.7, "edgecolor": "grey"}
title_specs     = {"fontsize": 14, "fontweight": "bold", "y": 1.2}
suptitle_specs  = {"fontsize": 11}
label_specs     = {"fontsize": 13}

plt.figure(figsize=(15,3))

plt.bar(x = features_sorted,
        height=tfidf_weights_sorted,
        **barplot_specs)

plt.title('Top words according to tf-idf in ingredients_en {}'.format(idx,df.iloc[idx,1]), **title_specs)
#plt.suptitle(df.iloc[idx,0],**suptitle_specs)
plt.ylabel('Tf-Idf weights', **label_specs)
plt.xticks(rotation=90)
plt.show()
```
