#### Helper Function to plot Outliers

```python
#Code by Tom C  https://www.kaggle.com/tjcdev/tps-outliers-hidden-features-baseline/notebook

def plot_outliers(df, feature, threshold=5):
    mean, std = np.mean(df), np.std(df)
    z_score = np.abs((df-mean) / std)
    good = z_score < threshold

    print(f"Rejection {(~good).sum()} points")
    visual_scatter = np.random.normal(size=df.size)
    plt.scatter(df[good], visual_scatter[good], s=2, label="Good", color="#4CAF50")
    plt.scatter(df[~good], visual_scatter[~good], s=8, label="Bad", color="#F44336")
    plt.legend(loc='upper right')
    plt.title(feature)
    plt.show();
    
    return good

def plot_lof_outliers(df, feature):
    lof = LocalOutlierFactor(n_neighbors=20, contamination=0.001, p=1)
    good = lof.fit_predict(df) > 0.5 # change this value to set the threshold for outliers
    print(f"Rejection {(~good).sum()} points")
    
    visual_scatter = np.random.normal(size=df.size)
    plt.scatter(df[good], visual_scatter[good], s=2, label="Good", color="#4CAF50")
    plt.scatter(df[~good], visual_scatter[~good], s=8, label="Bad", color="#F44336")
    plt.legend(loc='upper right')
    plt.title(feature)
    plt.show();
    
    return good
```

### Remove alone Outliers [> std x70 or (>std x35 and len<6)]

"Try to remove the records with extreme outliers which are over std x 70 from the mean, since it is not good to remove too many records. There are 26 records included in 5 indicators(columns)."

If we remove these 26 outlier records, it would improved our results when we run a ML model. (Number adapted to this task, check output below)

```python
feature_cols = ['carbohydrates_per_hundred', 'fat_per_hundred', 'fiber_per_hundred', 'protein_per_hundred', 'energy_per_hundred']
```

```python
#By Junji Takeshima https://www.kaggle.com/code/junjitakeshima/ubiquant-simple-lgbm-removing-outliers-en-jp

outlier_list = []
outlier_col = []

for col in feature_cols:
    
    temp_df = df1[(df1[col] > df1[col].mean() + df1[col].std() * 70) |
                       (df1[col] < df1[col].mean() - df1[col].std() * 70) ]
    temp2_df = df1[(df1[col] > df1[col].mean() + df1[col].std() * 35) |
                        (df1[col] < df1[col].mean() - df1[col].std() * 35) ]
    if len(temp_df) >0 : 
        outliers = temp_df.index.to_list()
        outlier_list.extend(outliers)
        outlier_col.append(col)
        print(col, len(temp_df))
    elif len(temp2_df)>0 and len(temp2_df) <6 :
        outliers = temp2_df.index.to_list()
        outlier_list.extend(outliers)
        outlier_col.append(col)
        print(col, len(temp2_df))

outlier_list = list(set(outlier_list))
print(len(outlier_col), len(outlier_list))
```
